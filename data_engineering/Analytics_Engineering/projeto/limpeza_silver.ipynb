```python
import pandas as pd
```

```python
data_path = 'data\\bronze\\'
```

```python
# Carregando os DataFrames
listings_path = data_path + 'listings_bronze.parquet'
calendar_path = data_path + 'calendar_bronze.parquet'
reviews_path = data_path + 'reviews_bronze.parquet'
listings_silver = pd.read_parquet(listings_path)
calendar_silver = pd.read_parquet(calendar_path)
reviews_silver = pd.read_parquet(reviews_path)
```

```python
listings_rows, listings_cols = listings_silver.shape
calendar_rows, calendar_cols = calendar_silver.shape
reviews_rows, reviews_cols = reviews_silver.shape
print(
    f'''
Listings: Linhas: {listings_rows}, Colunas {listings_cols}
Calendar: Linhas: {calendar_rows}, Colunas {calendar_cols}
Reviews: Linhas: {reviews_rows}, Colunas {reviews_cols}
    '''
)
```
