 #### 1. As músicas mais populares nos últimos 10 anos;

```python
most_popular = df[df['year'] > 2013].sort_values(by='popularity', ascending=False)
most_popular.head(10)
```

#### 2. A quantidade de músicas lançadas por artista e ano de lançamento;

```python
quant_musdic_artist = df.groupby('artists').size().sort_values(ascending=False).reset_index()
quant_musdic_artist.rename(columns={0:'Quantidade'}, inplace=True)
quant_musdic_artist.head()
```
