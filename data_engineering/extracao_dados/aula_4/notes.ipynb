```python
# IMPORTAR AS BIBLIOTECAS
import requests
import json 
import pandas as pd
```

```python
# URL PARA API (COM FILTRO APENAS DE DATA)
url = "https://olinda.bcb.gov.br/olinda/servico/taxaJuros/versao/v2/odata/TaxasJurosDiariaPorInicioPeriodo?$filter=InicioPeriodo ge '2024-09-01'"

```

```python
# REQUISIÇÃO PARA API
response = requests.get(url=url)
print(response)
```

```python
# TRANSFORMAR RESTPOSTA HTTP EM OBJETO JSON
dados = response.json()
dados
```

```python
# Criando o DataFrame no Pandas
df = pd.DataFrame(dados['value'])
df
```

```python
# Aleatórios
df.sample(10)
```

### ETL

```python
linhas, colunas = df.shape
f'{linhas} linhas e {colunas} colunas'
```

```python
##### Início de Transformação #####
# Verificar modalidades disponíveis

df['Modalidade'].unique()
```

```python
# Filtrar apenas modalidade 'Aquisição de veículos - Pré-Fixado'
df_veiculos = df[df['Modalidade'] == 'Aquisição de veículos - Pré-fixado']
df_veiculos.head()
```

```python

f'{df_veiculos.shape[0]} linhas e {df_veiculos.shape[1]} colunas'
```

```python
# Verificar Colunas codigoSegmento e codigoModalidade	

df_veiculos.isnull().sum()
```

```python
# Excluir colunas desnecessárias - Nulas
df_veiculos.drop(columns=['codigoSegmento', 'codigoModalidade'], inplace=True)
```

```python
df_veiculos.head()
```

```python
df_veiculos.dtypes
```

```python
# Modificando Tipo de coluna no DF

df_veiculos[['InicioPeriodo', 'FimPeriodo']] = df_veiculos[['InicioPeriodo', 'FimPeriodo']].astype('date32[pyarrow]')
```

```python
df_veiculos.dtypes
```

```python
df_veiculos.info(memory_usage='deep')
```

### CARGA 
Salvar alterações

```python
df_veiculos.to_csv('src/dados_veiculos.csv', sep=',', index=False) # csv
```
