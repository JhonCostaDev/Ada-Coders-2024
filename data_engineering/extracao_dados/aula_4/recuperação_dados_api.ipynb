### 1. COMO TER INFORMAÇÕES DAS TAXAS APLICADAS PELO MERCADO?

O Banco Central do Brasil disponibiliza através de uma API gratuita informações sobre taxa de juros para diversas modalidades de crédito.

https://dadosabertos.bcb.gov.br/dataset/taxas-de-juros-de-operacoes-de-credito/resource/cb682eb9-e9a9-497f-a1c4-299b385d09b9

### 2. COMO RECUPERAR ESSES DADOS?

Vamos utilizar a linguagem de programação Python para recuperar os dados


API <- **PYTHON** -> COLETA -> TRATAMENTO -> DISPONIBILIZAÇÃO

```python
# IMPORTAR BIBLIOTECAS
import requests
import json
import pandas as pd
```

```python
# URL PARA API (COM FILTRO APENAS DE DATA)
url = "https://olinda.bcb.gov.br/olinda/servico/taxaJuros/versao/v2/odata/TaxasJurosDiariaPorInicioPeriodo?$filter=InicioPeriodo ge '2024-09-01'"
```

```python
#REQUISIÇAO PARA API

resposta = requests.get(url)
print(resposta)
```

```python
#TRANSFORMAR RESPOSTA HTTP EM OBJETO JSON
dados = resposta.json()
dados
```

```python
#CRIANDO DATAFRAME PANDAS
df = pd.DataFrame(dados['value'])
df.head()
```
