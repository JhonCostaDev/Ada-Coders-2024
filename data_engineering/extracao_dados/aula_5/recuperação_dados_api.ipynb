### 1. COMO TER INFORMAÇÕES DAS TAXAS APLICADAS PELO MERCADO?

O Banco Central do Brasil disponibiliza através de uma API gratuita informações sobre taxa de juros para diversas modalidades de crédito.

https://dadosabertos.bcb.gov.br/dataset/taxas-de-juros-de-operacoes-de-credito/resource/cb682eb9-e9a9-497f-a1c4-299b385d09b9

### 2. COMO RECUPERAR ESSES DADOS?

Vamos utilizar a linguagem de programação Python para recuperar os dados


API <- **PYTHON** -> EXTRAÇÃO -> TRATAMENTO -> CARGA (ETL)
<br>
API <- **PYTHON** -> EXTRAÇÃO -> CARGA -> TRATAMENTO (ELT)

```python
# IMPORTAR BIBLIOTECAS
import requests
import json
import pandas as pd
```

```python
# URL PARA API (COM FILTRO APENAS DE DATA)
url = "https://olinda.bcb.gov.br/olinda/servico/taxaJuros/versao/v2/odata/TaxasJurosDiariaPorInicioPeriodo?$filter=InicioPeriodo ge '2024-09-01'"
```

```python
#REQUISIÇAO PARA API

resposta = requests.get(url)
print(resposta)
```

```python
#TRANSFORMAR RESPOSTA HTTP EM OBJETO JSON
dados = resposta.json()
```

```python
#CRIANDO DATAFRAME PANDAS
df = pd.DataFrame(dados['value'])
df.sample(10)
```

### ETL

```python
df.shape
```

```python
###### INÍCIO DE TRANSFORMAÇÃO ######

# verificar modalidades disponíveis

df['Modalidade'].unique()
```

```python
# filtrar apenas modalidade 'Aquisição de veículos - Pré-fixado'

df_veiculos = df[df['Modalidade'] == 'Aquisição de veículos - Pré-fixado'] 
df_veiculos.head()
```

```python
df_veiculos.shape
```

```python
# verificar colunas codigoSegmento e codigoModalidade

df_veiculos.isnull().sum() # verificar nulos em um DF
```

```python
# eliminado colunas nulas

df_veiculos.drop(columns=['codigoSegmento', 'codigoModalidade'], inplace=True)
df_veiculos.head()
```

```python
df_veiculos.dtypes
```

```python
# modificando tipo de coluna do DF
df_veiculos['InicioPeriodo'] = df_veiculos['InicioPeriodo'].astype('date32[pyarrow]')
df_veiculos['FimPeriodo'] = df_veiculos['FimPeriodo'].astype('date32[pyarrow]')
df_veiculos['Posicao'] = df_veiculos['Posicao'].astype('Int8')
```

```python
df_veiculos.info()
```

```python
df.head()
```

```python
#CARGA

df_veiculos.to_csv('src/etl/dados_veiculos.csv', sep=',', index=False) #csv
df_veiculos.to_parquet('src/etl/dador_veiculos.parquet', compression='snappy') #parquet
```

### ELT

```python
df.head()
```

```python
### CARGA DO ELT

df.to_csv('src/elt/dados_veiculos.csv', sep=',', index=False) # CSV
df.to_parquet('src/elt/dados_veiculos.parquet', compression='snappy')
```

```python
# TRANSFORMAÇÃO
df_veiculos = pd.read_parquet('src/elt/dados_veiculos.parquet')
df_veiculos.head()
```

```python
# FILTRAR MODALIDADE

df_veiculos_tratado = df_veiculos[df_veiculos['Modalidade'] == 'Aquisição de veículos - Pré-fixado']
df_veiculos_tratado.head()
```

```python
# EXCLUIR AS COLUNAS NULAS

df_veiculos_tratado.drop(columns=['codigoSegmento', 'codigoModalidade'], inplace=True)

# MODIFICAR TIPOS DE COLUNA
df_veiculos_tratado['InicioPeriodo'] = df_veiculos_tratado['InicioPeriodo'].astype('date32[pyarrow]')
df_veiculos_tratado['FimPeriodo'] = df_veiculos_tratado['FimPeriodo'].astype('date32[pyarrow]')
df_veiculos_tratado['Posicao'] = df_veiculos_tratado['Posicao'].astype('Int8')
```

```python
df_veiculos_tratado.head()
```

```python
df_veiculos_tratado.to_parquet('src/elt/sot/sot_dados_veiculos.parquet', compression='snappy')
```
