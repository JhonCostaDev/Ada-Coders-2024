```python
import pandas as pd 
import numpy as np 
df = pd.read_csv('data/penguins.csv')

df
```

```python
df.head()
```

```python
# Grafico de dispersão entre a grossura do bico e o comprimento do bico

df[['bill_length_mm', 'bill_depth_mm']]

df.plot(x='bill_length_mm', y='bill_depth_mm', kind='scatter')
```

```python
df.plot(x='bill_length_mm', y='bill_depth_mm', kind='scatter', s='flipper_length_mm')
```

```python
# Grafico com filtro

df[df['sex'] == 'FEMALE'].plot(x='bill_length_mm', y='bill_depth_mm', kind='scatter', title='Penguins Fêmeas')
```

```python
df[df['sex'] == 'MALE'].plot(x='bill_length_mm', y='bill_depth_mm', kind='scatter', title='Penguins Macho')
```

```python
ax1 = df[df['sex'] == 'MALE'].plot(x='bill_length_mm', y='bill_depth_mm', kind='scatter', title='Penguins Macho', color='blue')

ax2 = df[df['sex'] == 'FEMALE'].plot(x='bill_length_mm', y='bill_depth_mm', kind='scatter', title='Penguins Fêmeas', color='red')
```

```python
ax1 = df[df['sex'] == 'MALE'].plot(x='bill_length_mm', y='bill_depth_mm', kind='scatter', title='Penguins Macho', color='blue', label='Macho')


df[df['sex'] == 'FEMALE'].plot(x='bill_length_mm', 
                               y='bill_depth_mm', 
                               kind='scatter', 
                               title='Comprimento X Grossura do Bico', 
                               color='red', 
                               ax=ax1, 
                               label='Fêmea', 
                               xlabel='Comprimento do Bico', 
                               ylabel='Grossura do Bico')
```

## Gráficos com Barras



```python
df_vendas = pd.DataFrame({'Mês' : ['Jan', 'Fev', 'Mar', 'Abr', 'Jan', 'Fev', 'Mar', 'Abr', 'Jan', 'Fev', 'Mar', 'Abr'],
                         'Quantidade' : np.random.randint(low=100, size=12),
                          'Veículo' : np.repeat(['HB20', 'Onix', 'Sandero'],4)})
df_vendas
```

```python
df_vendas.groupby('Veículo').sum()
```

```python
df_vendas.groupby('Veículo').sum().plot(kind='bar')
```

```python
df_vendas.groupby('Veículo').sum().plot(kind='line')
```

```python
df_vendas.groupby('Mês').sum().plot(kind='bar', title='Vendas por Mês', xlabel = 'Mês', ylabel='Quantidade')
```
