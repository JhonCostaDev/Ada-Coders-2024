```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('data/Forbes_Richest_Atheletes_1990_2020.csv')

df
```

```python
# Contagem de nulos

df_nulos = df.isnull().sum()
df_nulos.plot(kind='bar', xlabel='Colunas', ylabel='Quantidade', color='green', title='Números de Nulos por Coluna')
```

```python
#.value_COunts em colunas categóricas
df_categoricas = df.select_dtypes(exclude=['number'])
df_categoricas.value_counts().reset_index()
```

```python
# projeto
lista_categoricas = df_categoricas.columns.to_list()

for coluna in lista_categoricas:
    re = df[coluna].value_counts()
    print(f'Coluna: {coluna}')
    print(re)
```

```python
#projeto
#.Describe() nas colunas numéricas
df_numericas = df.select_dtypes(include=['number'])
df_numericas.describe().plot(kind='bar')
```
