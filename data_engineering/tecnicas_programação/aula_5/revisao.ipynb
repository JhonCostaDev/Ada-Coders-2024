```python
# Aplicação de funções em um DataFrame
# Definir uma função para classificação de faltosos

def classifica_faltas(valor):
    if valor == 0:
        classificacao = 'Não Falta'
    elif valor > 3:
        classificacao = 'Falta Muito'
    else:
        classificacao = 'Falta Pouco'
    return classificacao
```

```python
# aplicar ao dataframe, a função percorre todos os valores da coluna falta

df['Faltas'].apply(classifica_faltas)
```

```python
# Criando uma nova coluna com essa classificação
df['classe_faltas'] = df['Faltas'].apply(classifica_faltas)
df
```

```python
# Value_counts para ver quantidade de faltantes
df['classe_faltas'].value_counts().reset_index()
```

```python
#Value_counts com parâmetro normalize=True para ter a proporção
df['classe_faltas'].value_counts(normalize=True).mul(100).round(2).reset_index()
```

```python
media_falta_muito = df[ df['classe_faltas'] ==  'Falta Muito']['Notas'].mean()
media_falta_pouco = df[ df['classe_faltas'] ==  'Falta Pouco']['Notas'].mean()
media_nao_falta = df[ df['classe_faltas'] ==  'Não Falta']['Notas'].mean()

print(f'''
Média de quem não falta: {media_nao_falta},
Média de quem falta pouco: {media_falta_pouco}
Média de quem falta muito: {media_falta_muito}
''')
```

```python
# Determinar a proporção de sobreviventes por classe

df_class = df_titanic.groupby(['Pclass', 'Survived']).size().reset_index()
df_class = df_class.rename(columns={'Pclass': 'Classe', 'Survived': 'Sobreviventes', 0:'Total'})
df_class
```

```python
# Rogerio
df_class_grp = df_titanic.groupby(['Pclass', 'Survived']).size().reset_index()
df_class_grp.rename(columns={0:'Quantidade'}, inplace=True) # efetiva as alterações
df_class_grp
```

```python
df_quantidade_total = df_titanic.groupby(['Pclass']).size().reset_index().rename(columns={0:'Quantidade Total'})
df_quantidade_total
```

#### Utilizando o JOIN, no pandas MERGE

```python
# Juntando tabelas com colunas em comum
df_sobreviventes = df_class_grp.merge(df_quantidade_total, on='Pclass', how='left')
df_sobreviventes
```

```python
# Calculando o percentual
df_sobreviventes['Percentual %'] = (df_sobreviventes['Quantidade'] / df_sobreviventes['Quantidade Total']).mul(100).round(2)

df_sobreviventes
```

```python
df_sobreviventes.sort_values('Percentual %')
```
